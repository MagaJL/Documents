```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/IgnacioPardo/Tecnologias_Exponenciales_2025/refs/heads/main/qatar.csv")
print(type(df))

# Ejercicios con graficos:

# SEGUNDA PARTE: Graficar los goles hechos por el equipo en cada partido

num_partidos = len(df)

fig, axes = plt.subplots(nrows=num_partidos, ncols=1, figsize=(6, num_partidos * 2))

if num_partidos == 1:
    axes = [axes]

for i in range(num_partidos):
    partido = df.iloc[i]
    selecciones = [partido['1'], partido['2']]
    goles = [partido['1_goals'], partido['2_goals']]

    ax = axes[i]
    ax.scatter(selecciones, goles, s=200, color=['royalblue', 'skyblue'])
    ax.set_title(f"Partido {i+1}: {selecciones[0]} vs {selecciones[1]}")
    ax.set_ylabel('Goles')
    ax.set_ylim(0, max(goles)+1)
    ax.grid(True, axis='y')

plt.tight_layout()
plt.show()



```
